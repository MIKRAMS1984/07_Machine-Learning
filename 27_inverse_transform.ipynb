```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
# impute missing values for age, embarked, embark_town and decK
df['age'].fillna(df['age'].median(), inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
df['deck'].fillna(df['deck'].mode()[0], inplace=True)
```

```python
df.drop('deck', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Dictionary to store the LabelEncoders for each column
label_encoders = {}

# Encode all categorical variables and object variables
for k in df.columns:
    if df[k].dtype == 'category' or df[k].dtype == 'object':
        le = LabelEncoder()
        df[k] = le.fit_transform(df[k])
        label_encoders[k] = le
df.head()

# Perform your operations (e.g., training a model, etc.)
```

```python
# Decode (inverse transform) the variables back to their original form
for k in df.columns:
    if k in label_encoders:
        le = label_encoders[k]
        df[k] = le.inverse_transform(df[k])

df.head()
```
