# Principal Component Analysis (PCA)

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()

pca = PCA(n_components=3)

iris_pca = pca.fit_transform(iris.data)

plt.scatter(iris_pca[:,0], iris_pca[:,1], c=iris.target)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()


plt.scatter(iris_pca[:,0], iris_pca[:,1], c=iris.target)
plt.xlabel('PCA2')
plt.ylabel('PCA3')
plt.show()
```

# PCA on TIPS dataset.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, LabelEncoder

tips = sns.load_dataset('tips')

le = LabelEncoder()
kat_features  =tips.select_dtypes(include='category').columns

for features in kat_features:
    tips[features] = le.fit_transform(tips[features])
    
scalar  = StandardScaler()
tips_scaled = scalar.fit_transform(tips)

pca = PCA()
x_pca = pca.fit_transform(tips_scaled)

plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of komponents')
plt.title('explained variase by komponents')
plt.ylabel('kumulative variase by komponents')
```
