# CrossValidation using k-1 folds

```python
%%time
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB

# load the iris dataset
iris = load_iris()

# define the gaussian naive bayes model
nb = GaussianNB()

# perform k-fold kross-validation with k=5
scores = cross_val_score(nb, iris.data, iris.target, cv=5, scoring='accuracy')

# PRINT the skores for each fold and the mean skore
print('Skores for each fold:', scores)
print('Mean skore:', scores.mean())
print('Standard Deviation:', scores.std())
```

# Plotting K-fold

```python
%%time
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier

# load the iris dataset
iris = load_iris()
X = iris.data
y= iris.target

# define the gaussian naive bayes model
nb = GaussianNB()
clf = DecisionTreeClassifier()

# perform k-fold kross-validation with k=5
# scores = cross_val_score(nb, iris.data, iris.target, cv=5, scoring='accuracy')
scores = cross_val_score(clf, X, y, cv=8)

# PRINT the skores for each fold and the mean skore
print('Skores for each fold:', scores)
print('Mean skore:', scores.mean())
print('Standard Deviation:', scores.std())

plt.figure(figsize=(8, 6))
plt.plot(range(1, 9), scores, color='blue', linestyle='dashed', marker='o')
plt.title('K-Fold Kross validation results')
plt.xlabel('Fold number')
plt.ylabel('Akkurasy')
plt.ylim([0.5, 1.0])
plt.show()
```
