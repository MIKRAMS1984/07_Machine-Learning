# Data Formatting

```python
mean = ks1['age'].mean()
mean
```

```python
ks1.info()
```

```python
ks1.describe()
```

```python
ks1['age'] = ks1['age'].replace(np.nan , mean)
ks1. isnull().sum().sort_values(ascending=False)
```

```python
ks1.dtypes
```

```python
ks1['survived'] = ks1['survived'].astype('int64')
ks1.dtypes
```

```python
ks1['age'] = ks1['age']*365
ks1.age.head(10)
```

```python
ks1.rename(columns={'age': 'age in days'}, inplace=True)
ks1.head()
```

```python
ks4 = ks1[['age in days', 'fare']]
ks4.head()
```
